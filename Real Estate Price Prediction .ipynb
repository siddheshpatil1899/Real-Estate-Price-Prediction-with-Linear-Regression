```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df5 = remove_pps_outliers(df4)
df5.shape
```

```python
def plot_scatter_chart(df,Location):
    no_of_bedrooms_2 = df[(df['Location'] == Location) & (df['No. of Bedrooms'] == 2)]
    no_of_bedrooms_3 = df[(df['Location'] == Location) & (df['No. of Bedrooms'] == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(no_of_bedrooms_2.Area,no_of_bedrooms_2.Price,color='red',label='2 BHK', s=50)
    plt.scatter(no_of_bedrooms_3.Area,no_of_bedrooms_3.Price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(Location)
    plt.legend()
    
plot_scatter_chart(df5,"Kharghar")
```

```python
plot_scatter_chart(df5,"Thane West")
```

```python
def remove_no_of_Bedrooms_outliers(df):
    exclude_indices = np.array([])
    for Location, Location_df in df.groupby('Location'):
        no_of_Bedrooms_stats = {}
        for no_of_Bedrooms, no_of_Bedrooms_df in Location_df.groupby('No. of Bedrooms'):
            no_of_Bedrooms_stats['No. of Bedrooms'] = {
                'mean': np.mean(no_of_Bedrooms_df.price_per_sqft),
                'std': np.std(no_of_Bedrooms_df.price_per_sqft),
                'count': no_of_Bedrooms_df.shape[0]
            }
        for no_of_Bedrooms, no_of_Bedrooms_df in Location_df.groupby('No. of Bedrooms'):
            stats = no_of_Bedrooms_stats.get(no_of_Bedrooms-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, no_of_Bedrooms_df[no_of_Bedrooms_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df6 = remove_no_of_Bedrooms_outliers(df5)
df6.shape
```

```python
dummies = pd.get_dummies(df6.Location)
dummies.head(3)
```

```python
df7 = pd.concat([df6,dummies.drop('other',axis='columns')],axis='columns')
df7.head()
```

```python
df8 = df7.drop('Location',axis='columns')
df8.head(3)
```

```python
df8.shape
```

```python
X = df8.drop(['Price'],axis='columns')
X.head(3)
```

```python
X.shape
```

```python
y = df8.Price
y.head(3)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
def predict_price(Location, Area, Car_Parking, no_of_Bedrooms):    
    loc_index = np.where(X.columns==Location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = Area
    x[1] = Car_Parking
    x[2] = no_of_Bedrooms
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price('Airoli',1000, 1, 2)
```

```python
predict_price('Kharghar',1000, 1, 2)
```
